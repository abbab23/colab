```
import pandas as pd
import streamlit as st
import numpy as np
from PIL import Image
import easyocr
import cv2
```

```
reader=easyocr.Reader(['en'])
def perform_ocr(image):
  image1=Image.open(image)
  image_np=np.array(image1)
  result=reader.readtext(image_np)
  return result
```

```
st.title('image to text')
uploaded_file=st.file_uploader("upload an image",type=['jpg','png','jpeg'])
if uploaded_file is not None:
  image=Image.open(uploaded_file)
  st.image(image,caption='uploaded image')
  st.write("performing ocr")
  result=perform_ocr(image)
  for (bbox,text,prob) in result:
     st.write(f"Detected text: {text} (Confidence: {prob:.2f})")
     image_np=np.array(image)
     cv2.rectangle(image_np, tuple(bbox[0]), tuple(bbox[2]), (0, 255, 0), 2)
  result_image=np.fromarray(image_np)
  st.image(result_image,caption='ocr result')
```

```
import pdfplumber
import torch
from transformers import pipeline
from sentence_transformers import SentenceTransformer
def extract_text(pdf):
  text=""
  with pdfplumber.open(pdf) as pdf:
    for page in pdf.pages:
      text+=page.extract_text()
      return text

extracted_text=extract_text("/content/INVOICE 1.pdf")
print(extracted_text)
```

```
pipeline=pipeline(model="deepset/roberta-base-squad2-distilled")
answer=pipeline(context=extracted_text,question="what is the invoice number and sales order no")
print(answer)
```

```
pipeline=pipeline("question-answering",model="deepset/roberta-base-squad2-distilled")
answer=pipeline(context=extracted_text,question="what is the invoice date")
print(answer)
```

```
from transformers import pipeline

# Load the question-answering pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Example extracted text (from a PDF or other source)
def extract_text(pdf):
  with pdfplumber.open(pdf) as pdf:
    text=""
    for page in pdf.pages:
      text=page.extract_text()
      return text
extracted_text=extract_text("/content/INVOICE (3).pdf")

# Define the question you want to ask
question = "what is the Sales Order No. "
question1 = "give me the Invoice No. "

# Perform question-answering
sales_order_no = qa_pipeline(question=question, context=extracted_text)
invoice_no = qa_pipeline(question=question1, context=extracted_text)

print(f"sales_order_no: {sales_order_no['answer']}")
print(f"invoice no: {invoice_no['answer']}")
```
